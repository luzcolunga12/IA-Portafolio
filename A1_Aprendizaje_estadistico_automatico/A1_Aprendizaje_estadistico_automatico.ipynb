```python
#se carga el archivo
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv('OneDrive/Documentos/UDEM/8 Semestre/Inteligencia Artificial I/Obesidad.csv')
df.head()
```

```python
#General desc
df.describe()
```

```python
#Separar por género mujer
mujeres= df[df['Sexo'] == 'Female']
mujeres.head()
```

```python
mujeres.describe()
```

```python
#Separar por género hombre
hombres= df[df['Sexo'] == 'Male']
hombres.head()
```

```python
hombres.describe()
```

```python
frecuencias = df["NivelDeObesidad"].value_counts()
print("Frecuencias absolutas:\n", frecuencias)
```

```python
porcentaje = frecuencias / len(df) * 100
print("Frecuencias absolutas:\n", porcentaje)
```

```python
frecuencias.plot()
plt.title("Distribución de Nivel de Obesidad") 
plt.xlabel("Categorías") 
plt.ylabel("Frecuencia")
plt.xticks(fontsize=5.5) 
plt.yticks(fontsize=8)
plt.show()
```

```python
import pandas as pd

tabla = pd.crosstab(df["NivelDeObesidad"], df["ComeMuchasCalorias"], normalize="columns") * 100
print(tabla.round(2))
```

```python
import matplotlib.pyplot as plt

tabla.plot(kind="bar", figsize=(8,5))
plt.title("Relación entre consumo de calorías y nivel de obesidad")
plt.xlabel("Nivel de Obesidad")
plt.ylabel("Porcentaje")
plt.legend(title="Come muchas calorías")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.pointplot(x="NivelDeObesidad", y="Peso", data=mujeres)
plt.title("Peso promedio por Nivel de Obesidad Mujeres")
plt.xlabel("Nivel de Obesidad")
plt.ylabel("Peso promedio (kg)")
plt.xticks(rotation=45)
plt.show()
```
